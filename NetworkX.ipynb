```python
import networkx as nx
import matplotlib.pyplot as plt
from itertools import combinations
```

```python
me = 'Alessio'
myNet = nx.Graph()

cyclomics = combinations([me,'Wigard', 'Jeroen'], 2)
myNet.add_edges_from(cyclomics)

bioinf = combinations([me, 'Mircea', 'Mark', 'Martin', 'Jose'], 2)
myNet.add_edges_from(bioinf)

cfDNA = combinations([me, 'Crhis', 'Alex', 'Christina'], 2)
myNet.add_edges_from(cfDNA)

students = combinations([me, 'Jeremy', 'Douwe'], 2)
myNet.add_edges_from(students)


plt.figure(dpi=180)
pos = nx.spring_layout(myNet, k=0.5,iterations=20)
nx.draw(myNet,
        pos         = pos,
        with_labels = True,
        font_weight = 'bold',
        font_size   = 8,
        node_size   = 2000,
        node_color  = 'limegreen')
```

```python
me = 'Alessio'
myNet = nx.Graph()


cyclomics = combinations([me,'Wigard', 'Jeroen'], 2)
myNet.add_edges_from(cyclomics)

bioinf = combinations([me, 'Mircea', 'Mark', 'Martin', 'Jose'], 2)
myNet.add_edges_from(bioinf)

cfDNA = combinations([me, 'Crhis', 'Alex', 'Christina', 'Ivo'], 2)
myNet.add_edges_from(cfDNA)

students = combinations([me, 'Jeremy', 'Douwe'], 2)
myNet.add_edges_from(students)

color_map = {'Alessio': 'lightcoral',
             'Wigard' : 'cyan',
             'Jeroen' : 'cyan',
             'Mircea' : 'limegreen',
             'Mark'   : 'limegreen',
             'Martin' : 'limegreen',
             'Jose'   : 'limegreen',
             'Crhis'  : 'orange',
             'Alex'   : 'orange',
             'Christina' : 'orange',
             'Ivo'    : 'orange',
             'Jeremy' : 'yellow',
             'Douwe'  : 'yellow'}

colors = [color_map[node] for node in myNet.nodes()]

        
plt.figure(dpi=200)
pos = nx.spring_layout(myNet, k=0.5,iterations=20)
nx.draw(myNet,
        pos         = pos,
        with_labels = True,
        font_weight = 'bold',
        font_size   = 8,
        node_size   = 2000,
        node_color  = colors)
```
