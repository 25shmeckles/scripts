```python
ref_date = datetime.datetime(2016,1,1,0,0,0)
periods = {'12h':'28days',
           '4h':'7days',
           '2h':'3days',
           '30min':'1day',
           '15min':'1day',
           '5min':'0.25days'}

dates = []
for k,v in periods.items():
    days = float(v.split('day')[0])
    start_date = ref_date - datetime.timedelta(days=days)
    data_range = pd.date_range(str(start_date),
                               str(ref_date),
                               freq=k,
                               closed='right').tolist()
    dates.append(data_range)
    
```

```python
for d in dates:
    print(len(d))
```
